```python
df=pd.read_csv("data_with_sent.csv")
df
```

```python
sentiment_count=[]
sentiment_pos=[]
sentiment_neg=[]
sentiment_neu=[]
#grouped by dates and sentiments
df.groupby(by=["Date"])["sentiment"].count()
#for unique days
date= df["Date"].unique()
for i in date:
    
    for r in range (0, df.shape[0]):
        #for positive counts
        p_count=0
        #for negative counts
        n_count=0
        #for neutral counts
        neu_count=0
        
        if (i==df.iloc[r]["Date"]):
            if(df.iloc[r]["sentiment"]=="Positive"):
                p_count=+1
            elif (df.iloc[r]["sentiment"]=="Negative"):
                n_count=+1
            else:
                neu_count=+1
      #positive sentiment list          
    sentiment_pos.append(p_count)
    #negative sentiment list
    sentiment_neg.append(n_count)
    #neutral sentiment list
    sentiment_neu.append(neu_count)
        
        
    
sentiment_pos  
sentiment_neg
sentiment_neu

                
            
            
    
   
    
```
